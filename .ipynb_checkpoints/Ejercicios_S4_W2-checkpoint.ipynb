# Let's code ⚡

```python
import pandas as pd
import numpy as np

# Generar datos de empleados
empleados = pd.DataFrame({
    'id_empleado': range(1, 11),
    'nombre': ['Ana', 'Luis', 'Carlos', 'María', 'Elena', 'Jorge', 'Clara', 'Pedro', 'Lucía', 'Andrés'],
    'id_departamento': np.random.choice([101, 102, 103, 104], size=10),
    'sueldo': np.random.randint(30000, 80000, size=10)
})

# Generar datos de departamentos
departamentos = pd.DataFrame({
    'id_departamento': [101, 102, 103, 104],
    'departamento': ['Ventas', 'Marketing', 'TI', 'Recursos Humanos']
})

# Generar datos de proyectos
proyectos = pd.DataFrame({
    'id_proyecto': range(1, 6),
    'nombre_proyecto': ['Alpha', 'Beta', 'Gamma', 'Delta', 'Epsilon'],
    'id_departamento': [101, 102, 103, 101, 104]
})

# Generar datos de participaciones en proyectos
participaciones = pd.DataFrame({
    'id_empleado': np.random.choice(range(1, 11), size=15, replace=True),
    'id_proyecto': np.random.choice(range(1, 6), size=15, replace=True)
})

# Generar un DataFrame con duplicados
duplicados = pd.DataFrame({
    'id_empleado': [1, 2, 2, 3, 3],
    'nombre': ['Ana', 'Luis', 'Luis', 'Carlos', 'Carlos'],
    'sueldo': [30000, 50000, 50000, 70000, 70000]
})

# Mostrar los datos generados
print("Datos de empleados:")
display(empleados)
print("\nDatos de departamentos:")
display(departamentos)
print("\nDatos de proyectos:")
display(proyectos)
print("\nDatos de participaciones:")
display(participaciones)
print("\nDatos con duplicados:")
display(duplicados)
```

### Ejercicio 3: Group by con `agg`

```python
df_grouped = empleados.groupby('id_departamento').agg(
    num_empleados=('id_empleado', 'count'),
    sueldo_promedio=('sueldo', 'mean'),
    sueldo_maximo=('sueldo', 'max')
).reset_index()

print("Resultado del group by con agregaciones:")
print(df_grouped)
```

### Ejercicio 4: Query para filtrar empleados

### Ejercicio 5: Eliminar duplicados

```python
# Solución
df_sin_duplicados = duplicados.drop_duplicates()
filas_eliminadas = len(duplicados) - len(df_sin_duplicados)

print("DataFrame sin duplicados:")
print(df_sin_duplicados)
print(f"Filas eliminadas: {filas_eliminadas}")
```

### Ejercicio 6: Merge con columnas diferentes


```python

df_merge1 = pd.merge(empleados, participaciones, on='id_empleado', how='inner')
df_merge_final = pd.merge(df_merge1, proyectos, on='id_proyecto', how='inner')

print("Relación de empleados con los proyectos en los que participan:")
print(df_merge_final[['nombre', 'nombre_proyecto', 'id_departamento']])
```

### Ejercicio 7: Concat con columnas distintas y agregaciones


```python
# Concatenar horizontalmente
df_concat_evaluaciones = pd.concat([empleados, evaluaciones['evaluacion_desempeño']], axis=1)

# Agrupar y realizar las agregaciones
df_agg = df_concat_evaluaciones.groupby('id_departamento').agg(
    promedio_evaluacion=('evaluacion_desempeño', 'mean'),
    total_empleados=('id_empleado', 'count')
).reset_index()

print("Promedio de evaluación y número de empleados evaluados por departamento:")
print(df_agg)
```

### Ejercicio 8: Query y eliminación de duplicados combinados

```python
# Solución
df_filtered = duplicados.query('sueldo > 50000 and nombre.str.startswith("C")', engine='python')
df_no_duplicates = df_filtered.drop_duplicates()

duplicados_eliminados = len(df_filtered) - len(df_no_duplicates)

print("Resultado del filtro y eliminación de duplicados:")
print(df_no_duplicates)
print(f"Duplicados eliminados: {duplicados_eliminados}")
```
