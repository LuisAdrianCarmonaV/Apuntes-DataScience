# Distribuciones de Probabilidad

## ¿Qué son las distribuciones de probabilidad?

Una distribución de probabilidad es una función que describe todas las posibles valores y probabilidades que una variable aleatoria puede tomar dentro de un rango específico. En términos simples, nos dice cuán probable es que ocurran ciertos resultados.

### Ejemplo del mundo real:
Imagina lanzar un dado justo. La distribución de probabilidad nos dirá que cada número del 1 al 6 tiene una probabilidad igual de:
$$
\text{Probabilidad} = \frac{\text{1}}{\text{6}}
$$.

(La probabiliad teorica se va a dar siempre y cuento el experimento se repita un número significativo de veces)

![mutuamente_excluyente_1](https://www.muelaner.com/wp-content/uploads/2013/07/central-limit-theorem.png)

---

## Tipos de Distribuciones de Probabilidad

### 1. **Distribuciones Discretas:**
   Estas distribuciones se aplican a variables aleatorias que pueden tomar un número finito de valores.

   - **Distribución Binomial**: Describe el número de éxitos en una secuencia de ensayos independientes y de igual probabilidad.
   - **Distribución de Poisson**: Describe el número de eventos que ocurren en un intervalo fijo de tiempo o espacio.

### 2. **Distribuciones Continuas:**
   Estas distribuciones se aplican a variables aleatorias que pueden tomar cualquier valor en un intervalo.

   - **Distribución Normal**: También conocida como la campana de Gauss, es una de las distribuciones más importantes en estadística.
   - **Distribución Uniforme**: Todos los valores en un rango tienen la misma probabilidad de ocurrir.


![mutuamente_excluyente_1](https://media.geeksforgeeks.org/wp-content/uploads/20240527193742/Probability-Distribution-01.webp)



---

## Distribuciones más usadas en Ciencia de Datos

1. **Distribución Normal (o Gaussiana):**
   - Frecuentemente utilizada para modelar fenómenos naturales como alturas, temperaturas, etc.
   - Es simétrica y su forma es de campana.

2. **Distribución Binomial:**
   - Utilizada para modelar experimentos de tipo éxito/fallo (como lanzar una moneda).

3. **Distribución de Poisson:**
   - Usada para modelar la cantidad de eventos que ocurren en un intervalo fijo de tiempo o espacio, como la cantidad de llamadas en un centro de soporte técnico.

4. **Distribución Uniforme:**
   - Se utiliza cuando todos los resultados posibles tienen la misma probabilidad de ocurrir.

## Visualización de Distribuciones

A continuación, veremos ejemplos de cómo crear distribuciones y visualizarlas usando Python.

## Distribución Binomial

### ¿Qué es la Distribución Binomial?

La distribución binomial es una distribución de probabilidad discreta que describe el número de éxitos en una secuencia de \( n \) ensayos independientes, cada uno con una probabilidad de éxito \( p \).

### Características:
- **Ensayos Independientes**: Cada ensayo no afecta el resultado de los demás.
- **Dos Resultados Posibles**: Éxito o fallo (por ejemplo, cara o cruz en un lanzamiento de moneda).
- **Número Fijo de Ensayos**: Se repite el experimento \( n \) veces.

### Ejemplo del mundo real:
Imagina que lanzas una moneda 10 veces. Quieres saber cuántas veces saldrá cara. Aquí, cada lanzamiento es un ensayo, y el número total de caras en 10 lanzamientos sigue una distribución binomial.

### Fórmula de la Distribución Binomial

La probabilidad de obtener exactamente \( k \) éxitos en \( n \) ensayos es:

$$
P(X = k) = \binom{n}{k} p^k (1 - p)^{n-k}
$$

donde:

- \( P(X = k) \) es la probabilidad de obtener exactamente \( k \) éxitos.
- \( n \) es el número total de ensayos.
- \( k \) es el número de éxitos deseados.
- \( p \) es la probabilidad de éxito en un ensayo.
- \( (1 - p) \) es la probabilidad de fracaso en un ensayo.
- \( \binom{n}{k} \) es el coeficiente binomial, que se calcula como:

$$
\binom{n}{k} = \frac{n!}{k!(n-k)!}
$$

### Parámetros Clave:
- \( n \): Número de ensayos.
- \( p \): Probabilidad de éxito en cada ensayo.

### Usos en Ciencia de Datos:
La distribución binomial es útil para modelar procesos donde se cuentan eventos binarios, como:
- Número de clics en una campaña de anuncios (clic o no clic).
- Número de productos defectuosos en una línea de producción.


## Ejemplo:Probabilidad de vida para pólizas de seguro

Un agente de seguros vende pólizas a cinco personas de la misma edad y que disfrutan de buena salud.
Según las tablas actuales, la probabilidad de que una persona en estas condiciones viva 30 años o más es 2/3.

Hállese la probabilidad de que, transcurridos 30 años, vivan:

1. Las cinco personas.

2. Exactamente dos personas.

3. Al menos tres personas.

```python
p = 2/3 #(la probabilidad de que una persona viva 30 años o más)
q = 1 - p #(la probabilidad de que una persona viva menos de 30 años)
n = 5  #(el número de personas a las que se les venden las pólizas)
k1 = 5

coeficiente_binomial_1 = factorial(n)/(factorial(k1)*factorial(n-k1))
prob_1 = round(
    coeficiente_binomial_1*(p**k1)*(q**(n-k1))
    ,3
)

prob_1
```

```python
k2 = 2

coeficiente_binomial_2 = factorial(n)/(factorial(k2)*factorial(n-k2))
prob_2 = round(
    coeficiente_binomial_1*(p**k2)*(q**(n-k2))
    ,3
)

prob_2
```

```python
def coef_binomial(n, k):
    return math.comb(n, k)

# Función para calcular la probabilidad binomial
def probabilidad_binomial(n, k, p):
    coef = coef_binomial(n, k)
    prob = coef * (p ** k) * ((1-p) ** (n - k))
    return prob

k3 = 3
k4 = 4
k5 = 5

# P(X >= 3) = P(X=3) + P(X=4) + P(X=5)
prob_3_o_mas_vivan = (
    probabilidad_binomial(n, k3, p) +
    probabilidad_binomial(n, k4, p) +
    probabilidad_binomial(n, k5, p)
)

prob_3_o_mas_vivan
```

### Implementacion en stats

```python
prob_5_vivan = stats.binom.pmf(5, n, p)

prob_2_vivan = stats.binom.pmf(2, n, p)

prob_3_o_mas_vivan = (
    stats.binom.pmf(3, n, p) +
    stats.binom.pmf(4, n, p) +
    stats.binom.pmf(5, n, p)
)

print(
    prob_5_vivan,
    prob_2_vivan,
    prob_3_o_mas_vivan
)
```

La razón de sumar las probabilidades es que, de acuerdo con las reglas de la probabilidad, cuando tenemos varios eventos mutuamente excluyentes (es decir, que no pueden ocurrir simultáneamente), la probabilidad de que ocurra uno de esos eventos es la suma de las probabilidades individuales.

## Distribución Normal

### ¿Qué es la Distribución Normal?

La distribución normal, también conocida como distribución de Gauss, es una distribución de probabilidad continua que es fundamental en estadística y ciencia de datos. Se caracteriza por su forma de campana simétrica, donde la mayoría de los valores se concentran alrededor de la media.

### Características:
- **Simetría**: La distribución es simétrica alrededor de la media.
- **Campana de Gauss**: La curva tiene forma de campana, donde los valores cerca de la media son los más comunes.
- **Media, Mediana y Moda Iguales**: En una distribución normal, la media, la mediana y la moda son iguales.


![mutuamente_excluyente_1](https://www.miraltabank.com/wp-content/uploads/2023/03/normal_distribution.png)

### Función de Densidad de Probabilidad (PDF):
La fórmula de la función de densidad para una variable aleatoria X con media mu y desviación estándar sigma es:

$$
f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{ -\frac{1}{2} \left( \frac{x - \mu}{\sigma} \right)^2 }
$$

Donde:
- mu es la media.
- sigma es la desviación estándar.

### Parámetros Clave:
- mu: Media de la distribución.
- sigma: Desviación estándar, que determina la dispersión de los datos.

### Usos en Ciencia de Datos:
- **Modelado de datos naturales**: Muchos fenómenos naturales se aproximan a una distribución normal.
- **Inferencia estadística**: Es fundamental en pruebas de hipótesis y cálculos de intervalos de confianza.

### Ejemplo del mundo real:
Alturas de personas, puntuaciones en exámenes, y errores de medición tienden a seguir una distribución normal.


```python
# Parámetros de la distribución normal
mu = 10  # Media
sigma = 1 # Desviación estándar

# Generar datos de la distribución normal
data_normal = np.random.normal(mu, sigma, 10000)

# Visualización de la distribución normal
sns.histplot(data_normal, kde=True, stat="density", linewidth=0)
plt.title('Distribución Normal')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.show()
```

| **Método**      | **Descripción**                                      | **Entrada**                          | **Salida**                                    | **Uso Común**                                   |
|-----------------|------------------------------------------------------|--------------------------------------|-----------------------------------------------|-------------------------------------------------|
| **`norm.cdf`**  | Calcula la probabilidad acumulada hasta un valor \(x\) | Valor \(x\), media \( \mu \), desviación \( \sigma \) | Probabilidad acumulada \( P(X \leq x) \)     | Obtener la probabilidad de \( X \leq x \)       |
| **`norm.ppf`**  | Calcula el valor \(x\) para una probabilidad acumulada \(p\) | Probabilidad \(p\), media \( \mu \), desviación \( \sigma \) | Valor \( x \) tal que \( P(X \leq x) = p \)   | Encontrar el valor \( x \) para una probabilidad dada |


### Ejercicio:

El tiempo que tarda en entregarse un pedido de una tienda en línea sigue una distribución normal con una media de 5 días y una desviación estándar de 1.5 días.

Encuentra la probabilidad de que un pedido seleccionado al azar tarde:

- Menos de 3.5 días en entregarse.
- Más de 7.5 días en entregarse.

![mutuamente_excluyente_1](https://raw.githubusercontent.com/anfagudelogo-tpt/Classes/refs/heads/main/imagenes/sprint_5_estadistica/Ejercicio1_normal.png)



```python
# Media y desviación estándar de la distribución
mean_delivery_time = 5
std_dev_delivery_time = 1.5

# Calcula la probabilidad de que el tiempo de entrega sea menor a 3.5 días
prob_less_than_3_5_days = norm.cdf(3.5, mean_delivery_time, std_dev_delivery_time)

# Calcula la probabilidad de que el tiempo de entrega sea mayor a 7 días
prob_more_than_7_days = 1 - norm.cdf(7.5, mean_delivery_time, std_dev_delivery_time)

prob_less_than_3_5_days, prob_more_than_7_days
```

### Ejercicio

Los precios de los pedidos realizados en una tienda virtual tienen una distribución normal con una media de 420 dólares y una desviación estándar de 65 dólares.

Algunos clientes eligen la entrega rápida por mensajería, que tiene un precio fijo independientemente del valor del pedido.

Los clientes tienden a molestarse cuando el costo de la entrega es igual al costo del pedido. ¿Cuánto debería costar el envío por mensajería para que no supere el precio del pedido en el 90% de los casos?

![mutuamente_excluyente_1](https://raw.githubusercontent.com/anfagudelogo-tpt/Classes/refs/heads/main/imagenes/sprint_5_estadistica/Ejercicio2_normal.png)




```python
mu = 420
sigma = 65
threshold = 0.90

max_delivery_price = norm(mu, sigma).ppf(1 - threshold)

print('Costo máximo de envío por mensajería:', max_delivery_price)
```

# Teorema del limite central

El Teorema del Límite Central establece que, independientemente de la distribución original de los datos, la distribución de la media de una muestra grande de datos tiende a ser normal (gaussiana) a medida que el tamaño de la muestra aumenta, siempre que la varianza sea finita. Este teorema es fundamental en estadística porque permite usar la distribución normal para hacer inferencias sobre medias muestrales, incluso si la distribución de los datos no es normal.

$$
Mu = n * p
$$

$$ Sigma = \sqrt{n \cdot p \cdot (1 - p)} $$
