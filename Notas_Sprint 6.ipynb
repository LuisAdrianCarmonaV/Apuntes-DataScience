## Diccionarios: recorrer diccionarios en bucle
Diccionarios: recorrer diccionarios en bucle

```python
financial_info = {
    'American Express': 93.23,
    'Boeing': 178.44,
    'Coca-Cola': 45.15,
    'Walt Disney': 119.34,
    'Nike': 97.99,
    'JPMorgan':96.27,
    'Walmart': 130.68 
}

for key in financial_info.keys():
    print(key)
```
