### Variables de shell vs. variables de entorno: cul es la diferencia?  

Es importante saber que hay dos tipos de "etiquetas con valor" en la Terminal:  

- Variables de shell: Estas son como notas adhesivas que solo puedes leer y usar dentro de la ventana de la Terminal donde las creaste. Otros programas que abras desde esa Terminal (como un Jupyter Notebook) no las vern automticamente.  
-  Variables de entorno: Estas son como notas adhesivas que se comparten con todos los programas que abres desde esa Terminal, incluyendo tus Jupyter Notebooks!

### Cmo crear tus propias variables (temporales):  

 (en maysculas, por convencin), seguido de un signo igual (=) y el valor que quieres guardar (si el valor tiene espacios, ponlo entre comillas):  

`MI_VARIABLE="mi primer valor"`  

`echo $MI_VARIABLE`  
Mostrar: `mi primer valor.`  

#### Cmo convertir una variable de shell en una variable de entorno (exportar)  

Para que tu variable est disponible para otros programas (como tus scripts de Python o Jupyter Notebooks), necesitas exportarla usando el comando export:  

`export MI_VARIABLE` 

Despus de hacer esto, `MI_VARIABLE` se convierte en una variable de entorno, y `printenv MI_VARIABLE` mostrar su valor.  

Tambin puedes crear y exportar una variable de entorno en un solo paso:  

`export OTRA_VARIABLE="otro valor"`
