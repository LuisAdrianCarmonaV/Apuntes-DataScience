##  Configurar un entorno virtual en Python paso a paso

### 1 Verificar qu versin de Python tienes instalada  
`python --version`   O usa `python3 --version` segn tu sistema

### 2 Crear un entorno virtual (llamado `mi_entorno`)  
`python -m venv mi_entorno`

### 3 Activar el entorno virtual  
**En Windows:**  
`mi_entorno\Scripts\activate` 
**En macOS/Linux:**  
`source mi_entorno/bin/activate`

### 4 Instalar paquetes dentro del entorno virtual  
`pip install pandas numpy matplotlib`

### 5 Ver los paquetes instalados  
`pip list`

### 6 Desactivar el entorno virtual cuando termines  
`deactivate`

## Controlar las versiones de tus "piezas": precisin!  
`conda install --name mi_entorno scipy=0.15.0`  
`conda create --name mi_entorno python=3.9  `  

### Compartir la "lista de piezas" de tu "mini-taller": archivos de entorno  

Cuando trabajas en equipo, es crucial que todos tengan exactamente las mismas "piezas" y versiones. Puedes guardar la "lista de piezas" de tu entorno en un archivo (normalmente llamado `requirements.txt`) y compartirlo con otros.  

Para crear un archivo con la lista de paquetes de tu entorno actual:  

`conda list --export > requirements.txt`  

Esto crear un archivo requirements.txt con los nombres de las libreras y sus versiones exactas. Aqu tienes un ejemplo de cmo se vera este archivo:  

`pandas==1.3.1`  
`scipy==1.6.2`  
`streamlit==1.12.2`  
Para crear un entorno virtual a partir de este archivo:  

`conda env create --name mi_nuevo_entorno --file requirements.txt`
