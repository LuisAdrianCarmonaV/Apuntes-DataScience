# üöÄ Subir carpeta local a GitHub desde VS Code o Git Bash

Este cuaderno resume los pasos completos para conectar una carpeta local (como tus apuntes o notebooks) a un repositorio de GitHub y subirla correctamente.

## üß≠ 1. Verificar instalaci√≥n de Git

Abre Git Bash o la terminal de VS Code y ejecuta:
```bash
git --version
```
Si muestra algo como `git version 2.x.x`, ya lo tienes instalado.

## ‚öôÔ∏è 2. Configurar usuario y correo global (solo la primera vez)
Reemplaza con tu nombre y correo de GitHub:
```bash
git config --global user.name "Luis Adrian Carmona Villalobos"
git config --global user.email "luisadrian.carmona@outlook.com"
```
Verifica que se guard√≥ correctamente:
```bash
git config --global --list
```

## üìÅ 3. Inicializar el repositorio local
Ub√≠cate en la carpeta de tu proyecto (por ejemplo `Apuntes`) y ejecuta:
```bash
cd "ruta/a/tu/carpeta/Apuntes"
git init
```
Esto crea una carpeta oculta `.git` que convierte tu proyecto en un repositorio local.

## üßπ 4. Eliminar carpetas problem√°ticas (opcional)
Si ves un error como `error: 'Sprint_7/test_render/' does not have a commit checked out`, b√≥rrala:
```bash
rm -rf "Sprint_7/test_render/"
```

## üì¶ 5. A√±adir archivos al repositorio
```bash
git add .
```
El punto (`.`) significa que a√±adir√° todos los archivos del proyecto.

## üìù 6. Crear el primer commit
```bash
git commit -m "Primer commit con apuntes y notebooks"
```

## üåø 7. Cambiar el nombre de la rama principal a main
```bash
git branch -M main
```

## ‚òÅÔ∏è 8. Conectar con el repositorio de GitHub
Usa tu URL del repositorio:
```bash
git remote add origin https://github.com/LuisAdrianCarmonaV/Apuntes-DataScience.git
```

## üöÄ 9. Subir los archivos a GitHub
```bash
git push -u origin main
```
La primera vez puede pedirte iniciar sesi√≥n o un token personal (PAT).

## üîí 10. (Opcional) Crear un `.gitignore`
Para evitar subir carpetas como `.ipynb_checkpoints` o `__pycache__`, crea un archivo llamado `.gitignore` en la ra√≠z de tu carpeta con el siguiente contenido:
```bash
# Carpetas de Jupyter
.ipynb_checkpoints/

# Archivos temporales de VS Code
.vscode/

# Archivos de cach√© de Python
__pycache__/
*.pyc

# Carpeta de pruebas
test_render/
```
Luego actualiza el repositorio:
```bash
git add .gitignore
git commit -m "Agregado .gitignore para ignorar archivos temporales"
git push
```
