```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Datos de ejemplo
data = {
    "Producto": ["A", "B", "C", "D"],
    "Ventas": [150, 200, 300, 100],
    "Devoluciones": [20, 30, 50, 10]
}
df = pd.DataFrame(data)
df
```

# Gráficos de Barras

Los gráficos de barras son una herramienta común para representar datos categóricos. Cada barra representa una categoría, y su altura o longitud indica la magnitud de esa categoría en relación con otras.

## ¿Cuándo usar gráficos de barras?
- Comparar valores entre diferentes categorías (ejemplo: ventas por producto, calificaciones por materia).
- Visualizar datos discretos y categóricos.
- Mostrar tendencias en categorías nominales u ordinales.

## Características principales
1. **Ejes**:
   - El eje X representa las categorías.
   - El eje Y muestra la escala de los valores.

2. **Tipos de gráficos de barras**:
   - **Barras verticales**: Uso típico.
   - **Barras horizontales**: Útiles cuando los nombres de las categorías son largos o hay muchas categorías.
   - **Barras apiladas**: Comparan subcategorías dentro de una misma categoría.
   
3. **Particularidades**:
   - En `matplotlib`, usamos `plt.bar()` para barras verticales y `plt.barh()` para horizontales.
   - En `seaborn`, la función principal es `sns.barplot()`, que permite agregar automáticamente medidas estadísticas como medias o medianas.

## Casos de uso
1. Comparar ventas de productos.
2. Analizar distribuciones de calificaciones por clase.
3. Comparar tasas de desempleo entre países.


```python
# Gráfico de barras con Matplotlib
plt.figure(figsize=(10, 5))
plt.bar(df["Producto"], df["Ventas"], color='skyblue', label="Ventas [Cantidad]")
plt.bar(df["Producto"], df["Devoluciones"], color='salmon', label="Devoluciones [Cantidad]")
plt.title("Ventas y Devoluciones por Producto")
plt.xlabel("Producto")
plt.ylabel("Cantidad")
plt.legend()
plt.show()
```

```python
test = df.melt(id_vars="Producto", var_name="Tipo", value_name="Cantidad")

test
```

```python
# Gráfico de barras con Seaborn
plt.figure(figsize=(10, 5))
sns.barplot(data=test,  x="Producto",y="Cantidad",hue="Tipo",palette="muted",)
plt.title("Ventas y Devoluciones por Producto (Seaborn)")
plt.show()
```

# Histogramas

Un histograma es una representación gráfica de la distribución de un conjunto de datos continuos. Divide los datos en intervalos (llamados *bins*) y muestra la frecuencia de valores en cada intervalo. A diferencia de los gráficos de barras, los histogramas están diseñados para representar datos numéricos y no categóricos.

## ¿Cuándo usar histogramas?
- Analizar la distribución de datos (ejemplo: edades, ingresos, puntuaciones en un examen).
- Identificar patrones como sesgos, concentraciones o distribuciones uniformes.
- Detectar valores atípicos o anomalías.

## Características principales
1. **Ejes**:
   - El eje X representa los intervalos (*bins*).
   - El eje Y muestra la frecuencia de los datos dentro de cada intervalo.

2. **Personalización**:
   - Número de *bins*: Afecta la granularidad del gráfico. Pocos *bins* pueden ocultar detalles, demasiados pueden generar ruido.
   - Rango de valores: Se puede limitar el rango para enfocar áreas específicas.

3. **En librerías**:
   - En `matplotlib`, usamos `plt.hist()`.
   - En `seaborn`, utilizamos `sns.histplot()`, que también permite combinar histogramas con estimaciones de densidad (*KDE*).

## Casos de uso
1. Analizar la distribución de edades en una población.
2. Visualizar puntuaciones en un examen para evaluar rendimiento.
3. Entender patrones de ingresos en diferentes regiones.


```python
# Generar datos de ejemplo
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)  # Distribución normal con media=50 y desviación estándar=10
```

```python
# Histograma con Matplotlib
plt.figure(figsize=(10, 5))
plt.hist(data, color='skyblue', edgecolor='black', alpha=1, bins=30)
plt.title("Distribución de Datos (Matplotlib)")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.show()
```

```python
# Histograma con Seaborn
plt.figure(figsize=(10, 5))
sns.histplot(data, kde=True, color='teal',stat = 'count')
plt.title("Distribución de Datos con KDE (Seaborn)")
plt.xlabel("Valor")
# plt.ylabel("Frecuencia")
plt.show()
```

# Gráficos de Dispersión (Scatterplot)

Un gráfico de dispersión es ideal para visualizar la relación entre dos variables numéricas. Cada punto en el gráfico representa una observación, con su posición determinada por los valores de las dos variables.

## ¿Cuándo usar gráficos de dispersión?
- Identificar relaciones o patrones entre dos variables (ejemplo: altura vs. peso).
- Visualizar correlaciones positivas, negativas o la ausencia de ellas.
- Detectar valores atípicos o grupos en los datos.

## Características principales
1. **Ejes**:
   - El eje X representa una variable independiente.
   - El eje Y representa una variable dependiente.

2. **Personalización**:
   - Tamaño y color de los puntos: Se pueden usar para agregar información adicional, como una tercera variable.
   - Estilo: Se pueden agregar líneas de tendencia para analizar relaciones.

3. **En librerías**:
   - En `matplotlib`, usamos `plt.scatter()` para crear gráficos básicos.
   - En `seaborn`, utilizamos `sns.scatterplot()`, que permite agregar más dimensiones mediante colores, tamaños o estilos.

## Casos de uso
1. Analizar cómo la temperatura afecta las ventas de helados.
2. Relacionar horas de estudio con calificaciones.
3. Evaluar cómo el ingreso influye en el gasto.

**Nota**: Un gráfico de dispersión es más útil cuando ambas variables son cuantitativas y continuas.

```python
# Generar datos de ejemplo
np.random.seed(42)
n = 100
data_1 = {
    "Horas de Estudio": np.random.uniform(1, 10, n),
    "Calificaciones": np.random.uniform(50, 100, n) + np.random.uniform(-10, 10, n),
    "Horas de Sueño": np.random.uniform(5, 9, n)
}
data_2 = pd.DataFrame(data_1)
data_2
```

```python
# Gráfico de dispersión con Matplotlib
plt.figure(figsize=(10, 5))
plt.scatter(data_2["Horas de Estudio"], data_2["Calificaciones"])
plt.title("Relación entre Horas de Estudio y Calificaciones")
plt.xlabel("Horas de Estudio")
plt.ylabel("Calificaciones")
# plt.colorbar(label="Horas de Sueño")
plt.grid(alpha=0.5)
plt.show()
```

```python
# Gráfico de dispersión con Seaborn
plt.figure(figsize=(10, 5))
sns.scatterplot(
    data=data_2,
    x="Horas de Estudio",
    y="Calificaciones",
    hue="Horas de Sueño",
    size="Horas de Sueño",
    sizes=(20, 200),
    palette="viridis"
  )
plt.title("Relación entre Horas de Estudio, Calificaciones y Horas de Sueño")
plt.xlabel("Horas de Estudio")
plt.ylabel("Calificaciones")
plt.legend(title="Horas de Sueño", loc="upper left", bbox_to_anchor=(1, 1))
plt.show()
```

# Gráficos de Línea (Lineplot)

Un gráfico de línea se utiliza para mostrar cómo cambia una variable a lo largo de una dimensión continua, como el tiempo. Es ideal para analizar tendencias, patrones y fluctuaciones.

## ¿Cuándo usar gráficos de línea?
- Visualizar tendencias a lo largo del tiempo (ejemplo: ventas por mes, temperatura diaria).
- Mostrar la evolución de datos secuenciales.
- Comparar series temporales o patrones entre grupos.

## Características principales
1. **Ejes**:
   - El eje X generalmente representa la variable continua (como tiempo).
   - El eje Y muestra los valores de la variable de interés.

2. **Personalización**:
   - Estilo de líneas: Colores, tipos de líneas (sólida, punteada) y grosores.
   - Marcadores: Agregar puntos para enfatizar datos específicos.
   - Subgrupos: Usar diferentes colores o estilos para comparar múltiples series.

3. **En librerías**:
   - En `matplotlib`, usamos `plt.plot()` para gráficos básicos de líneas.
   - En `seaborn`, utilizamos `sns.lineplot()` para gráficos más complejos que permiten agregar dimensiones adicionales como subgrupos o bandas de confianza.

## Casos de uso
1. Analizar la tendencia de ventas durante el año.
2. Visualizar el cambio de temperatura a lo largo del día.
3. Comparar la evolución de la población entre países.

**Nota**: Los gráficos de línea son efectivos cuando las observaciones están ordenadas en una dimensión continua.


```python
# Generar datos de ejemplo
np.random.seed(42)
dias = pd.date_range(start="2024-01-01", periods=30)
ventas = np.random.uniform(50, 150, len(dias))
devoluciones = np.random.uniform(5, 30, len(dias))

data_3 = pd.DataFrame({
    "Fecha": dias,
    "Ventas": ventas,
    "Devoluciones": devoluciones
})
data_3.head()
```

```python
# Gráfico de línea con Matplotlib
plt.figure(figsize=(10, 5))
plt.plot(data_3["Fecha"], data_3["Ventas"], label="Ventas", color="blue", marker="o")
plt.plot(data_3["Fecha"], data_3["Devoluciones"], label="Devoluciones", color="red", linestyle="--", marker="x")
plt.title("Ventas y Devoluciones Diarias")
plt.xlabel("Fecha")
plt.ylabel("Cantidad")
plt.xticks(rotation=45)
plt.legend()
plt.grid(alpha=0.5)
plt.show()
```

```python
# Gráfico de línea con Seaborn
plt.figure(figsize=(10, 5))
sns.lineplot(data=data_3, x="Fecha", y="Ventas", label="Ventas", marker="o", color="blue")
sns.lineplot(data=data_3, x="Fecha", y="Devoluciones", label="Devoluciones", marker="x", color="red", linestyle="--")
plt.title("Ventas y Devoluciones Diarias (Seaborn)")
plt.xlabel("Fecha")
plt.ylabel("Cantidad")
plt.xticks(rotation=45)
plt.legend()
plt.grid(alpha=0.5)
plt.show()
```

```python
data_3.head()
```

```python
data_3.melt(id_vars="Fecha", var_name="Tipo", value_name="Cantidad").head()
```

```python
# Gráfico de línea con Seaborn
plt.figure(figsize=(10, 5))
sns.lineplot(
    data=data_3.melt(id_vars="Fecha", var_name="Tipo", value_name="Cantidad"),
    x="Fecha",
    y="Cantidad",
    hue='Tipo',
   )
plt.title("Ventas y Devoluciones Diarias (Seaborn)")
plt.xlabel("Fecha")
plt.ylabel("Cantidad")
plt.xticks(rotation=45)
plt.legend()
plt.grid(alpha=0.5)
plt.show()
```

# Gráfico de Cajas y Bigotes (Boxplot)

El gráfico de cajas y bigotes (boxplot) es una excelente herramienta para visualizar la distribución de un conjunto de datos, identificar valores atípicos (outliers) y comparar diferentes grupos. Muestra los cuartiles de un conjunto de datos y destaca los valores atípicos.

## ¿Cuándo usar un boxplot?
- Analizar la distribución de datos en términos de mediana, cuartiles y rangos.
- Identificar posibles valores atípicos o extremos en los datos.
- Comparar distribuciones entre varios grupos o categorías.

## Características principales
1. **Caja**: Representa el rango intercuartílico (IQR), que va desde el primer cuartil (Q1) hasta el tercer cuartil (Q3).
2. **Línea en el centro de la caja**: Representa la mediana (Q2).
3. **Bigotes**: Se extienden desde los cuartiles hasta el valor máximo o mínimo dentro de 1.5 veces el IQR desde los cuartiles.
4. **Valores atípicos**: Puntos fuera de los bigotes que se consideran outliers.

## Casos de uso
1. Comparar distribuciones de ingresos entre diferentes grupos.
2. Analizar la dispersión y los valores atípicos en puntuaciones de exámenes.
3. Visualizar la variabilidad de una variable en diferentes categorías.

**Nota**: El boxplot es muy útil para resumir grandes cantidades de datos y comparar distribuciones de manera visual.


```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Generar datos de ejemplo
np.random.seed(42)
grupo_A = np.random.normal(100, 15, 200)  # Media = 100, Desviación estándar = 15
grupo_B = np.random.normal(110, 20, 200)  # Media = 110, Desviación estándar = 20
grupo_C = np.random.normal(95, 10, 200)   # Media = 95, Desviación estándar = 10

df_4 = pd.DataFrame({
    "Grupo A": grupo_A,
    "Grupo B": grupo_B,
    "Grupo C": grupo_C
})
```

```python
pd.Series(grupo_A).describe()
```

```python
# Boxplot con Matplotlib
plt.figure(figsize=(10, 5))
plt.boxplot(grupo_A, labels=["Grupo A"],)
plt.title("Distribución de Datos por Grupo (Matplotlib)")
plt.ylabel("Valor")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

```python
# Boxplot con Seaborn
plt.figure(figsize=(10, 5))
sns.boxplot(data=df_4.melt(),y = 'value', hue='variable', palette="Set2")
plt.title("Distribución de Datos por Grupo (Seaborn)")
plt.ylabel("Valor")
plt.show()
```

# Gráfico de Violín (Violin Plot)

El gráfico de violín combina características de los gráficos de cajas y los histogramas. Muestra la distribución de los datos a través de un formato similar a un violín, permitiendo visualizar la densidad de los datos, así como los cuartiles, la mediana y los valores atípicos.

## ¿Cuándo usar un violin plot?
- Visualizar la distribución de datos y la densidad en un solo gráfico.
- Comparar distribuciones de varios grupos o categorías.
- Identificar simetría o sesgo en los datos.

## Características principales
1. **Distribución**: El cuerpo del violín muestra la densidad de probabilidad de los datos, lo que permite ver la forma de la distribución.
2. **Caja interna**: Similar al boxplot, se muestra la mediana y el rango intercuartílico (IQR).
3. **Bigotes**: También se incluyen los valores máximos y mínimos dentro de 1.5 veces el IQR, similar al boxplot.
4. **Ancho**: El ancho del "violín" en diferentes puntos indica la densidad de los datos en esos valores.

## Casos de uso
1. Comparar distribuciones de calificaciones en diferentes grupos de estudiantes.
2. Visualizar la variabilidad en el ingreso de diferentes países o regiones.
3. Explorar la forma y densidad de distribuciones de datos.

**Nota**: El gráfico de violín es particularmente útil cuando se desea ver tanto la distribución como la densidad de los datos al mismo tiempo.


```python
# Generar datos de ejemplo
np.random.seed(42)
grupo_A = np.random.normal(100, 15, 200)  # Media = 100, Desviación estándar = 15
grupo_B = np.random.normal(110, 20, 200)  # Media = 110, Desviación estándar = 20
grupo_C = np.random.normal(95, 10, 200)   # Media = 95, Desviación estándar = 10

df_5 = pd.DataFrame({
    "Grupo A": grupo_A,
    "Grupo B": grupo_B,
    "Grupo C": grupo_C
})
```

```python
df_5.melt()
```

```python
# Violin plot con Seaborn
plt.figure(figsize=(10, 6))
sns.violinplot(data=df_5.melt(),x='value',hue='variable', inner="box", palette="muted")
plt.title("Distribución de Datos por Grupo (Violin Plot)")
plt.ylabel("Valor")
plt.xlabel("Grupo")
plt.show()
```
