# **C칩mo combinar DataFrames con `pandas.concat`**

En an치lisis de datos, muchas veces necesitamos combinar datos que est치n separados en varias tablas (DataFrames). Una de las herramientas m치s 칰tiles para esto es la funci칩n `pandas.concat`.

## **쯈u칠 es `pandas.concat`?**

La funci칩n `pandas.concat` permite unir **DataFrames** o **Series** de manera sencilla. Puedes usarla para:
1. **Combinar filas**: Apilar DataFrames uno debajo del otro.
2. **Combinar columnas**: Unir DataFrames uno al lado del otro.

### **Par치metros principales**

- `objs`: Una lista de DataFrames o Series que quieres combinar.
- `axis`: Define c칩mo se combinan los datos:
  - `axis=0` (por defecto): Une filas (apila los DataFrames verticalmente).
  - `axis=1`: Une columnas (pega los DataFrames horizontalmente).
- `ignore_index`: Si es `True`, ignora los 칤ndices originales y asigna nuevos 칤ndices.
- `keys`: Asigna etiquetas a cada DataFrame en la concatenaci칩n.

### **Ejemplos comunes**

1. Combinar dos DataFrames con el mismo n칰mero de columnas (apilado vertical).
2. Combinar dos DataFrames con diferentes columnas, agregando `NaN` donde faltan valores.
3. Usar etiquetas para identificar el origen de los datos.

Con `pandas.concat`, tienes flexibilidad para trabajar con estructuras de datos m치s complejas de forma eficiente. 춰Vamos a practicar! 游꿢


```python
# Ejercicio 1: Combina dos DataFrames verticalmente
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})

display(
    df1,
    df2
)
```

```python
# Combina df1 y df2 apil치ndolos uno debajo del otro
df_vertical = pd.concat(
    [df1, df2],
    axis= 0,
    ignore_index=True
    )
print("DataFrame combinado verticalmente:")
df_vertical
```

```python
df3 = pd.DataFrame({'C': [9, 10]})
df3
```

```python
# Ejercicio 2: Combina dos DataFrames horizontalmente

df_horizontal = pd.concat([df1, df3], axis=1)
print("\nDataFrame combinado horizontalmente:")
df_horizontal
```

```python
# Ejercicio 3: Usa etiquetas para identificar el origen de los datos
df_with_keys = pd.concat([df1, df2], keys=['Datos1', 'Datos2'])
print("\nDataFrame con etiquetas:")
df_with_keys
```

# **Combinar DataFrames con `pandas.merge`**

Cuando trabajas con datos, es com칰n que necesites combinar informaci칩n de varias tablas bas치ndote en alguna columna o clave en com칰n. La funci칩n `pandas.merge` te permite hacer esto, de manera similar a las operaciones de **JOIN** en bases de datos.

## **쯈u칠 es `pandas.merge`?**

La funci칩n `pandas.merge` combina dos DataFrames en funci칩n de una o m치s columnas que compartan. Es ideal para unir datos relacionados de una forma precisa y flexible.

### **Tipos de combinaciones**

- **Inner Join** (por defecto): Solo conserva las filas donde las claves coincidan en ambos DataFrames.
- **Left Join**: Conserva todas las filas del DataFrame izquierdo y las coincide con las del derecho.
- **Right Join**: Conserva todas las filas del DataFrame derecho y las combina con las del izquierdo.
- **Outer Join**: Conserva todas las filas de ambos DataFrames y rellena con `NaN` donde no haya coincidencias.

### **Par치metros principales**

- `how`: Tipo de combinaci칩n. Puede ser `'inner'`, `'left'`, `'right'` o `'outer'`.
- `on`: Columna(s) comunes en ambos DataFrames para realizar la combinaci칩n.
- `left_on` y `right_on`: Especifican columnas diferentes si los nombres no coinciden.
- `suffixes`: A침ade sufijos para las columnas con nombres duplicados.

### Ejemplo t칤pico

Imagina que tienes un DataFrame con informaci칩n de empleados y otro con datos de los departamentos. Puedes usar `merge` para combinar esta informaci칩n bas치ndote en una columna como `id_empleado` o `id_departamento`.

춰Vamos a practicar con ejemplos para entender c칩mo funciona! 游꿢


```python
empleados = pd.DataFrame({
    'id_empleado': [1, 2, 3,4],
    'nombre': ['Ana', 'Luis', 'Carlos','Andres'],
    'id_departamento': [101, 102, 101,110]
})

departamentos = pd.DataFrame({
    'id_departamento': [101, 102, 103],
    'departamento': ['Ventas', 'Marketing', 'Recursos Humanos'],
    'nombre':['Ventas', 'Marketing', 'Recursos Humanos']
})

display(
    empleados,
    departamentos
)
```

```python
# Ejercicio 1: Combina los DataFrames con un Inner Join
df_inner = empleados.merge(
    departamentos,
    on='id_departamento',
    how='inner',
    suffixes = ['_emp','_dept']
    )
print("Inner Join:")
df_inner
```

```python
# Ejercicio 2: Combina los DataFrames con un Left Join
df_left = empleados.merge(
    departamentos,
    on='id_departamento',
    how='left')
print("\nLeft Join:")
df_left
```

```python
# Ejercicio 3: Combina los DataFrames con un Outer Join
df_outer = pd.merge(
    empleados,
    departamentos,
    on=['id_departamento','asdasd','asdasd'],
    how='outer')
print("\nOuter Join:")
df_outer
```

```python
empleados_renombrado = empleados.rename(columns={'id_departamento': 'departamento_id'})

empleados_renombrado
```

```python
df_different_columns = empleados_renombrado.merge(
    departamentos,
    left_on='departamento_id',
    right_on='id_departamento',
    how='inner'
)

df_different_columns
```
